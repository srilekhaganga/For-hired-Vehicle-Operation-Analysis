```python
# API Endpoint
api_url = "https://data.cityofnewyork.us/resource/8wbx-tsch.json"
limit = 1000  # Maximum limit per request
offset = 0  # Starting offset
total_records = 0  # Counter

while True:
    # Construct the API request with limit and offset
    paginated_url = f"{api_url}?$limit={limit}&$offset={offset}"

    # Fetch data from API
    response = requests.get(paginated_url)

    if response.status_code == 200:
        data = response.json()  # Convert response to JSON

        # If no more data, stop
        if not data:
            print("No more data to fetch.")
            break

        # Insert data into MongoDB
        collection.insert_many(data)
        total_records += len(data)

        print(f"Inserted {len(data)} records from offset {offset}. Total: {total_records}")

        # Move to the next batch
        offset += limit

        # Avoid API rate limits
        time.sleep(1)
    else:
        print(f"Error fetching data: {response.status_code}, Response: {response.text}")
        break

print(f"Total records inserted: {total_records}")
```
